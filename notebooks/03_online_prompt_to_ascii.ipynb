# üåê Online Prompt-to-ASCII Art Generator

**No 30GB download required!** Uses HuggingFace's free cloud API.

## Setup
1. Get a free API token at: https://huggingface.co/settings/tokens
2. Paste it below and run!

```python
# üîë PASTE YOUR FREE API KEY HERE
HF_TOKEN = "hf_your_token_here"  # Get free at huggingface.co/settings/tokens

# Setup
import sys
sys.path.insert(0, '..')

from ascii_gen.online_generator import OnlineGenerator
from ascii_gen.production_training import ProductionCNNMapper
from IPython.display import display

# Initialize
print("üöÄ Setting up...")
gen = OnlineGenerator(api_key=HF_TOKEN)

# Load CNN mapper
cnn = ProductionCNNMapper()
try:
    cnn.load("../models/production_cnn.pth")
except:
    print("Training CNN mapper...")
    cnn.train(epochs=50)

print("‚úÖ Ready! Enter prompts below.")
```

---
## ‚úèÔ∏è Enter Your Prompt

```python
# üé® YOUR PROMPT HERE:
prompt = "A majestic mountain with snow peaks, simple line art style"

# Generate image via API
print(f"üé® Generating: {prompt}")
image = gen.generate(prompt, width=512, height=384)

if image:
    print("\nüì∑ Generated Image:")
    display(image)
    
    # Convert to ASCII
    ascii_art = cnn.convert_image(image)
    print("\nüé≠ ASCII Art:")
    print("=" * 60)
    print(ascii_art)
    print("=" * 60)
else:
    print("‚ùå Generation failed. Check your API key.")
```
