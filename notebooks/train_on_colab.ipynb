```python
# 4. Upload your Config and Dataset
# Please upload the 'ascii_training_data' folder from your local machine to your Google Drive root.
# Also upload 'config/train_flux_ascii.yaml' to Drive.

import os
import shutil

# Copy config to local workspace
if not os.path.exists("config.yaml"):
    shutil.copy("/content/drive/MyDrive/train_flux_ascii.yaml", "config.yaml")

# Update config to point to Drive dataset
# We will replace the dataset path in the yaml
with open("config.yaml", "r") as f:
    content = f.read()

content = content.replace("/workspace/ascii_training_data", "/content/drive/MyDrive/ascii_training_data")

with open("config.yaml", "w") as f:
    f.write(content)

print("âœ… Config updated to use Drive dataset")
```
